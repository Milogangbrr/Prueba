```python
# Importar las librerías necesarias
import pandas as pd
import json
import os
from IPython.display import display  # Importar display para mostrar el DataFrame de manera más interactiva

# Configurar pandas para mostrar todas las columnas
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
pd.set_option('display.max_rows', None)     # Mostrar todas las filas (ten cuidado si el DataFrame es muy grande)
pd.set_option('display.width', None)        # Ajustar automáticamente el ancho de las columnas
pd.set_option('display.max_colwidth', None) # No recortar el contenido de las celdas

# Ruta para el archivo JSON (modificar si es necesario)
json_folder = os.path.join(os.getcwd(), '..', 'json')  # Asegúrate de que la carpeta json esté en el directorio adecuado
json_file_path = os.path.join(json_folder, 'series_enero_2024.json')

# Leer el archivo JSON
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Convertir los datos JSON a un DataFrame
df = pd.json_normalize(data)

# Mostrar el DataFrame completo
display(df)  # Usamos display() para mostrar de manera interactiva

# Verificación de tipos de datos y estructura del DataFrame
display(df.info())  # También mostramos la información de la estructura del DataFrame
```
